# Making the AWS AMI for mumax-ec2 #
Author: Colin Jermain <clj72@cornell.edu>

The Amazon Machine Image (AMI) used for mumax-ec2 is created for Amazon Web Service (AWS).

## Starting with the NVIDIA AMI ##
AWS has an image made for Amazon Linux that includes the latest CUDA drivers. At the time of this writing the CUDA version is 6.5 (AMI released 10/16/2014), which is required by MuMax 3.6. The NVIDIA AMI is used as the base for the mumax-ec2 AMI so that the latest CUDA is inhereted. The latest version of the NVIDIA AMI should be used in the following process to upgrade the CUDA version.

1. Visit the [NVIDIA AMI](https://aws.amazon.com/marketplace/pp/B00FYCDDTE)
2. Click "Continue"
3. Click "Manual Launch"
4. "Launch with EC2 Console" under the prefered region
5. Choose the GPU instance (g2.2xlarge at the time of this writing)
6. Continue with default settings until step (5) Tag Instance
7. Add Key: `mumax-ec2`  Value: `1.0`  (where `1.0` is the version number of the mumax-ec2 AMI)
8. Choose an appropriate security group
9. Create a security keypair and save the `pem` file to a safe place (`~/.ssh/`)
10. Change the permissions of the `pem` file for read access only for the current user

```bash
cd ~/.ssh
chmod 600 mumax-ec2.pem
```

11. Launch the instance: i-b4e40645
12. Get the public IP address: 54.165.25.230
13. Add the connection details to the `~/.ssh/config/` file

```
Host mumax-ec2
HostName 54.165.25.230
User ec2-user
IdentityFile "~/.ssh/mumax-ec2.pem"
```

14. Connect to the instance

```bash
ssh mumax-ec2
```

15. Accept the authenticity of the host
16. Update the packages

```bash
sudo yum update
```

17. Download the latest version of MuMax

```bash
wget https://github.com/mumax/3/releases/download/v3.6.1/mumax3.6.1_linux_x64.tar.lzma
```

18. Extract the MuMax archive

```bash
tar --lzma -xvf mumax3.6.1_linux_x64.tar.lzma
```

19. Install `git`

```bash
sudo yum install git
```

20. Get the latest version of mumax-ec2

```bash
git clone https://github.com/ralph-group/mumax-ec2.git
```
